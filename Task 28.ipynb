# Data Types in NumPy

```python
import numpy as np 
arr = np.array([1, 2, 3, 4])
print(arr.dtype)
```

```python
arr = np.array(['apple', 'banana', 'cherry'])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 3, 4], dtype='S')
print(arr)
print(arr.dtype)  
'''The b prefix: This stands for "bytes."
S means "String."
1 means "1 byte long" '''
```

```python
arr=np.array([1,2,3,4,5],dtype='i4')
print(arr)
print(arr.dtype)
```

# What if a Value Can Not Be Converted?

```python
arr=np.array(['a','2','3'])
print(arr)
arr[0]="0"
new=arr.astype('i')
print(new.dtype)
```

```python
arr = np.array([1.1, 2.1, 3.1])
new=arr.astype(int)
print(new.dtype)
```

```python
arr = np.array([-1, 0, 10])
new=arr.astype(bool)
print(new)
print(new.dtype)
```
